```python
#Check for missing values ​​for each feature.
import missingno as msno
msno.bar(data)
```

```python
#Check what percentage of missing values ​​are included.
missing_data = data.isnull().sum()
total = data.count()
missing_ratio = total/len(data)

ratio = {'Total':total, 'missing_ratio' : missing_ratio}
print(pd.DataFrame(ratio))
```

```python
#Delete features (columns) that have 5 or more missing values.
print(data.shape)
for col in data.columns:
  if data[col].isnull().sum() > 5:
    data.drop(col, axis=1, inplace = True)

#63 columns left
print("After deleted:")
print(data.shape)
```

```python
#Samples (rows) with missing values ​​are deleted from the data from which features with 5 or more missing values ​​have been deleted.
#data1 = data
#print (data1.isnull().sum())
print(data.shape)
data.dropna(axis = 0, thresh=5, inplace=True)
print("After deleted:")
print(data.shape)
```

[Problem 5]

```python
# Terminology survey
#kurtosis
#Kurtosis is a measure of the combined weight of a distribution's tails relative to the center of the distribution. When a set of approximately normal data is graphed via a histogram, it shows a bell peak and most data within three standard deviations (plus or minus) of the mean.
print("kurtosis")
kurt = pd.DataFrame(data=data).kurt(axis=0)
# e.g. SalePrice and LotArea kurtosis are over 3 which are leptokurtic distribution 
# MSSubClass is platykurtic distribution which is kurtosis < 3
print(kurt)

#Skewness
#Skewness is a measure of the asymmetry of a distribution. A distribution is asymmetrical when its left and right side are not mirror images. A distribution can have right (or positive), left (or negative), or zero skewness.
print("skewness")
skew =  pd.DataFrame(data=data).skew(axis=0)
# e.g. SalePrice and LotArea are highly skewed 
print (skew)
```

[Problem 6]

```python
import seaborn as sns
sns.displot(data['SalePrice'], kde=True).set(title='SalePrice plot kurt = ' + str(kurt['SalePrice']) + ' and skew = ' + str(skew['SalePrice']))
print('SalePrice is leptokurtic distribution kurtosis > 3')

sns.displot(data['MSSubClass'], kde=True).set(title='MSSubClass plot kurt = ' + str(kurt['MSSubClass']) + ' and skew = ' + str(skew['MSSubClass']))
print('MSSubClass is platykurtic distribution kurtosis < 3')
```

[Problem 7]

```python
#Make a heat map of the correlation coefficient matrix and find a relationship with a high correlation coefficient.
corr = data.corr(method = 'pearson').abs()
print(corr)

sns.heatmap(corr).set(title = 'Correlation')
```

```python
#@title Default title text
#Select 10 features that are highly correlated with the objective variable, and create a correlation coefficient matrix heat map for these.

corr_high = corr.where(corr>0.3)
corr1 = corr_high.dropna(axis = 1, thresh = 10)
corr2 = data[corr1.columns].corr()
sns.heatmap(corr2).set(title = 'Highly correlated 10 features')

print ('Highly correlated 10 features are: ' + str(corr1.columns))
```

```python
# High correlated 3 features

corr_higher = corr2.where(corr2>0.64)
corr3 = corr_higher.dropna(axis = 1, thresh = 3)
print ('High correlated 3 features are: ' + str(corr3.columns))
```
