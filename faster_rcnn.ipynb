[Problem 2] Code reading

Most important part of Faster RCNN is Region Proposal Networks (RPNs) that are designed to efficiently predict region proposals with a wide range of scales and aspect ratios. In the training region, RPN takes the feature map as input and outputs region proposals. RPN has both a classifier and a regressor, where the classifier gives the probability of the region containing the object and the regressor gives the coordinates of the bounding boxes. And these proposals go to the ROI pooling layer. ROI pooling layer reshapes the region proposals into fixed-size feature maps, and then used to classify the image within the proposed region and predict the offset values for the bounding boxes.

[Problem 4] Create a file for learning

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
annotation_df = pd.read_csv('/content/drive/Othercomputers/My_Laptop/Google_Drive/ObjectDetection-master/annotation.txt',header=None)
```

```python
print (annotation_df.head())
```

```python
le = LabelEncoder()
annotation_df.iloc[:,5] = le.fit_transform(annotation_df.iloc[:,5])
```

```python
rcnn_path = '/content/drive/Othercomputers/My_Laptop/Google_Drive/ObjectDetection-master/annotation.txt'
yolo_path = 'annotation.txt'
n_sample, n_col = annotation_df.shape
with open(rcnn_path) as f:
    for i in range(n_sample):
        lines = f.readline() 
        split_line = lines.split(',')
        #print (split_line)
        image_path = split_line[0]
        split_line[0] = '../ObjectDetection-master/' + image_path
        split_line[-1] = str(annotation_df.iloc[i,5]) + '\n'  
        #print (split_line)
        with open(yolo_path, mode='a') as out_f:
            join_line = ','.join(split_line)  
            join_line = join_line.replace('.jpg,','.jpg ')  
            out_f.write(join_line)
```
