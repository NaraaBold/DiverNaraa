```python
# Make a pie chart of the number of samples for each label (display up to the percentage)
import matplotlib.pyplot as plt

#count samples for each label
count_samples = df.groupby('Species').count()

labels = list(data.target_names)
sizes = count_samples.iloc[:,0].tolist()

plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```

```python
#Select features one by one and visualize the distribution of data for each label using a box plot.
#Select features one by one and visualize the distribution of data for each label using a violin plot
import matplotlib.patches as mpatches

legend1 = []

def box_plot(ax, data, edge_color, fill_color):
    bp = ax.boxplot(data, patch_artist=True)
    
    for element in ['boxes', 'whiskers', 'fliers', 'means', 'medians', 'caps']:
        plt.setp(bp[element], color=edge_color)

    for patch in bp['boxes']:
        patch.set(facecolor=fill_color)       
        
    return bp

legends = []

def violin_plot(ax, data, label):
    bp = ax.violinplot(data)
    
    color = bp["bodies"][0].get_facecolor().flatten()
    legends.append((mpatches.Patch(color=color), label))
        
    return bp

fig1, axs = plt.subplots(nrows=1, ncols=2, figsize=(18, 8))
axs[0].set_title('Box Plot')
axs[1].set_title('Violin Plot')
data1  = [df[df['Species']==0].iloc[:, 0], df[df['Species']==0].iloc[:, 1], df[df['Species']==0].iloc[:, 2], df[df['Species']==0].iloc[:, 3]] # 0th species
data2  = [df[df['Species']==1].iloc[:, 0], df[df['Species']==1].iloc[:, 1], df[df['Species']==1].iloc[:, 2], df[df['Species']==1].iloc[:, 3]] # 1st species
data3  = [df[df['Species']==2].iloc[:, 0], df[df['Species']==2].iloc[:, 1], df[df['Species']==2].iloc[:, 2], df[df['Species']==2].iloc[:, 3]] # 2nd species
# draw box plot
d1 = box_plot(axs[0], data1, 'red', 'tan')
d2 = box_plot(axs[0], data2, 'blue', 'cyan')
d3 = box_plot(axs[0], data3, 'green', 'beige')
# draw violin plot
d4 = violin_plot(axs[1], data1, labels[0])
d5 = violin_plot(axs[1], data2, labels[1])
d6 = violin_plot(axs[1], data3, labels[2])

for ax in axs:
    ax.yaxis.grid(True)
    ax.set_xticks([1,2,3,4],["sepal_length", "sepal_width", "petal_length", "petal_width"])
    ax.set_xlabel('Features')
    ax.set_ylabel('cm')

plt.setp(axs, xticks=[1,2,3,4],
         xticklabels=["sepal_length", "sepal_width", "petal_length", "petal_width"])

axs[1].legend(*zip(*legends), loc=1)
axs[0].plot([], c='red', label=labels[0])
axs[0].plot([], c='blue', label=labels[1])
axs[0].plot([], c='green', label=labels[2])
axs[0].legend()
plt.show()

```
